```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
quarterly = pd.read_csv('quarterly.csv', index_col=0, header=[0], parse_dates=True, date_parser=pd.to_datetime)
quarterly.head()
```

```python
plt.plot(quarterly)
```

```python
# https://stackoverflow.com/questions/17001389/pandas-resample-documentation
```

```python
upsampled = quarterly.resample('MS').interpolate(method='linear')
upsampled.head(35)
```

```python
plt.plot(upsampled)
```

```python
print(quarterly.shape[0])
print(upsampled.shape[0])
```

```python
upsampled.to_csv('upsampled.csv')
```
