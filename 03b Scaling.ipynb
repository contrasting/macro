```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# we will apply Z-normalisation
```

```python
full = pd.read_csv('growth_rate.csv', index_col=0, header=[0], parse_dates=True)
full.head()
```

```python
full.shape
```

```python
normalised = pd.DataFrame()

for series in full:
    mean = full[series].mean()
    std = full[series].std()
    normalised[series] = (full[series] - mean)/std
```

```python
normalised.head()
```

```python
normalised.shape
```

```python
plt.plot(normalised)
```

```python
normalised.to_csv('normalised_growth.csv')
```
